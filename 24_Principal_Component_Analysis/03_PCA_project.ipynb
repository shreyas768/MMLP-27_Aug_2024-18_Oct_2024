## Scaling Data

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
sc_pixels = sc.fit_transform(pixels)
```

### Perform PCA

```python
from sklearn.decomposition import PCA
```

```python
pca2 = PCA(n_components=2)

pca_pixels = pca2.fit_transform(sc_pixels)
pca_pixels
```

```python
np.sum(pca2.explained_variance_ratio_)
```

```python
explained_variance = []

for n in range(1,65):
    pca = PCA(n_components=n)
    pca.fit(sc_pixels)
    
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1,65),explained_variance)
plt.xlabel('No of Components')
plt.ylabel('Variance Explained')
plt.ylim(0.4,1.0)
plt.show()
```

```python
# Scatterplot using 2 PCAs
plt.figure(figsize=(10,6))
sns.scatterplot(x=pca_pixels[:,0],y=pca_pixels[:,1],hue = df['number_label'],palette = 'Set1')
plt.legend(loc=(1.05,0))
```
