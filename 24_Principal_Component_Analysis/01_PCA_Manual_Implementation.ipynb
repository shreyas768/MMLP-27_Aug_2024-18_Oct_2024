```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cancer_tumor_data_features.csv')
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc_X = sc.fit_transform(df)
sc_X
```

```python
sc_X.mean()
```

```python
sc_X = sc_X - sc_X.mean()
sc_X
```

```python
#Grab Covariance Matrix
covariance_matrix = np.cov(sc_X,rowvar=False)
```

```python
#Get Eigen Vectors and Eigen Values

eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
#Choosing number of components
num_components = 2
```

```python
#Get index sorting key based on Eigen values
sorted_key = np.argsort(eigen_values)[::-1][:num_components]
```

```python
#Get num_components of Eigen Values and Eigen Vectors
eigen_values, eigen_vectors = eigen_values[sorted_key],eigen_vectors[:,sorted_key]
```

```python
eigen_values
```

```python
eigen_vectors
```

```python
#dot product of original data and eigen_vectors are the PC values
#This is the projection of original points to the PCAs

prinicipal_components = np.dot(sc_X,eigen_vectors)
prinicipal_components
```

```python
plt.scatter(prinicipal_components[:,0],prinicipal_components[:,1])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer

cancer_df = load_breast_cancer()
```

```python
cancer_df['target']
```

```python
plt.scatter(prinicipal_components[:,0],prinicipal_components[:,1], c=cancer_df['target'])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```
