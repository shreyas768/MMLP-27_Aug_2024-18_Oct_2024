```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img_array = mpimg.imread('palm_trees.jpg')  #RBG codes for each pixel
img_array.shape
```

```python
plt.figure(figsize=(8,8))
plt.imshow(img_array)
plt.show()
```

## Use KMeans to quantize Colors

```python
(h,w,c)=img_array.shape
```

```python
h
```

```python
w
```

```python
c
```

```python
img_array_2d = img_array.reshape(h*w,c)
```

```python
img_array_2d.shape
```

```python
from sklearn.cluster import KMeans
```

```python
mod = KMeans(n_clusters=15)
```

```python
labels = mod.fit_predict(img_array_2d)
labels
```

```python
np.unique(labels)
```

```python
rgb_codes = mod.cluster_centers_.round(0).astype(int)
```

```python
quantized_image = np.reshape(rgb_codes[labels],(h,w,c))
```

```python
plt.figure(figsize=(6,6))
plt.imshow(quantized_image)
plt.show()
```
