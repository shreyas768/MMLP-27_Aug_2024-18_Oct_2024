```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.info()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='sales',y='salary',data=df,hue='division')
```
