```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('StudentsPerformance.csv')
df.head(3)
```

```python
#jointplot

sns.jointplot(x='math score',y='reading score',data=df,hue='gender')
```

```python
sns.jointplot(x='math score',y='reading score',data=df,kind='reg')
```

```python
sns.jointplot(x='math score',y='reading score',data=df,kind='kde')
```

```python
#pairplot

sns.pairplot(df)
```

```python
sns.pairplot(df,hue='gender',corner=True,diag_kind='hist')
```
