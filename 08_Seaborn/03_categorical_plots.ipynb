```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head(3)
```

```python
#countplot: no of rows per category
plt.figure(figsize=(10,3))
sns.countplot(x='level of education',data=df)
```

```python
plt.figure(figsize=(10,3))
sns.countplot(y='level of education',data=df,hue='division',orient='h')
```

```python
#barplot
sns.barplot(x='level of education',y='salary',data=df, estimator= np.mean)
```

```python
#boxplot
df1 = pd.read_csv('StudentsPerformance.csv')
df1.head()
```

```python
df1.info()
```

```python
#boxplot
plt.figure(figsize=(10,6))
sns.boxplot(x='parental level of education',y='math score', data=df1, hue='gender',width=0.5)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(y='parental level of education',x='math score', data=df1, hue='gender', orient='h')
plt.yticks(rotation=45)
plt.show()
```

```python
#violinplot
sns.violinplot(x='parental level of education',y='writing score', data=df1)
```

```python
sns.violinplot(x='parental level of education',y='writing score', data=df1, hue='gender',split=True,inner = 'quartiles')
```

```python
#swarmplot
sns.swarmplot(x='reading score',data=df1,size=2)
```

```python
sns.swarmplot(x='reading score',y='race/ethnicity',data=df1,size=3.7,hue='gender')
```
