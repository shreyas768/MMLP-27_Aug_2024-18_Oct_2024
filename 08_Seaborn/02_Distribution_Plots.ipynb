```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head(3)
```

```python
#Rugplot
sns.rugplot(x='sales',data=df,height=0.5)
```

```python
#Histogram
sns.set_theme(style='whitegrid')
sns.histplot(data=df,x='salary',bins=10,kde=True)
plt.show()
```

```python
sns.kdeplot(data=df,x='salary')
```

```python
sns.kdeplot(data=df,x='sales',y='salary')
```

```python
sns.scatterplot(data=df,x='sales',y='salary',alpha=0.5)
```
