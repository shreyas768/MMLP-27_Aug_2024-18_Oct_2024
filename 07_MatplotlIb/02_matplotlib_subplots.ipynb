```python
import numpy as np
import matplotlib.pyplot as plt

a = np.linspace(0,10,11)
b = a**4
x = np.arange(0,10)
y = 2*x
```

```python
fig, axes = plt.subplots()

axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(b,a)
axes[1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=2,figsize=(12,7))

axes[0,0].plot(a,b)
axes[0,0].set_xlabel('this is a')


axes[1,1].plot(b,a)
axes[1,1].set_title('b vs a')

axes[0,1].plot(x,y)
axes[1,0].plot(y,x)

fig.suptitle('Figure Level title here')

plt.tight_layout()
plt.savefig('example.png')
plt.show()
```

### Styling

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label = 'x squared')
ax.plot(x,x**3, label='x cubed')
ax.legend(loc = 4)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label = 'x squared', color = '#752D2D',marker = '8',ls = '--')
ax.plot(x,x**3, label='x cubed', color = 'black', ls = '-.',lw=7)
ax.legend(loc = (1.1,0.5))
```
