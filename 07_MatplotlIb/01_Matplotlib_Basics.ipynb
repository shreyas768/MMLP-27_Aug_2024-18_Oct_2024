MATLAB in python

Pros: very easy, custom labels and texts, great control over each and every aspect of figure

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#basic array plot

x = np.arange(0,10)
x
```

```python
y = 2*x
y
```

```python
#plt.plot()

plt.figure(figsize=(6,2))
plt.plot(x,y)
plt.xlabel('This is X axis title')
plt.ylabel('This is Y axis title')
plt.title('Title of the plot')
plt.xlim(1,5)
plt.ylim(2.5,7.5)
plt.savefig('example.png')
plt.show() #similar to print statement
```

### Matplotlib Object Oriented Method

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
a
```

```python
b
```

```python
x
```

```python
y
```

```python
#create figure (empty canvas)
fig = plt.figure()

#add a set of axes to figure
axes = fig.add_axes([0,0,1,1]) #left,bottom,width,height (range 0 to 1)


#plot on these axes
axes.plot(x,y)

plt.savefig('example.png')
plt.show()
```

### Multiple Plots

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1]) #larger fig
axes2 = fig.add_axes([0.2,0.2,0.5,0.5]) #smaller fig

axes1.plot(a,b)
axes1.set_xlabel('This is a')
axes1.set_ylabel('This is b')
axes1.set_title('This is Title')
axes1.set_xlim(2,8)

axes2.plot(x,y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1]) #larger fig
axes2 = fig.add_axes([0.2,0.2,0.5,0.5]) #smaller fig
axes3 = fig.add_axes([1,1,0.25,0.45])

axes1.plot(a,b)
axes1.set_xlabel('This is a')
axes1.set_ylabel('This is b')
axes1.set_title('This is Title')
axes1.set_xlim(2,8)

axes2.plot(x,y)

axes3.plot(b,a)

fig.savefig('example.png',bbox_inches='tight')
```
