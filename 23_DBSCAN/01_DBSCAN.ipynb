```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
blobs = pd.read_csv('cluster_blobs.csv')
blobs.head()
```

```python
sns.scatterplot(data=blobs, x='X1',y='X2')
```

```python
moons = pd.read_csv('cluster_moons.csv')
moons.head()
```

```python
sns.scatterplot(data=moons, x='X1',y='X2')
```

```python
circles = pd.read_csv('cluster_circles.csv')
circles.head()
```

```python
sns.scatterplot(data=circles,x='X1',y='X2')
```

```python
#Label Discovery Function

def display_categories(model,data1):
    labels = model.fit_predict(data1)
    sns.scatterplot(data=data1, x='X1',y='X2', hue = labels, palette='Set1')
```

```python
#KMeans

from sklearn.cluster import KMeans

mod = KMeans(n_clusters=2)
display_categories(mod,moons)
```

```python
display_categories(mod,circles)
```

```python
mod = KMeans(n_clusters=3)
display_categories(mod,blobs)
```

## DBSCAN Results

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.6)
```

```python
display_categories(db,blobs)
```

```python
db = DBSCAN(eps=2)
display_categories(db,blobs)
```

```python
db = DBSCAN(eps=0.2)
display_categories(db,moons)
```

```python
db = DBSCAN(eps=0.2)
display_categories(db,circles)
```
