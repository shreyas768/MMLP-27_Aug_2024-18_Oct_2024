```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
df.head()
```

```python
df.info()
```

```python
def last_four(num):
    return int(str(num)[-4:])
```

```python
last_four(8757886)
```

```python
last_four(df['CC Number'][0])
```

```python
#apply
df['CC Number'] = df['CC Number'].apply(last_four)
```

```python
df.head()
```

```python
df['tip'].mean()
```

```python
def cust_rate(t1):
    if t1 < 1.5:
        return '*'
    elif t1>=1.5 and t1<3.5:
        return '**'
    else:
        return '***'
    
df['ratings'] = df['tip'].apply(cust_rate)
df.head(3)
```

```python
df['total_bill'].apply(lambda n:1.05*n)
```

```python
def quality(bill,t1):
    if t1/bill > 0.25:
        return 'Generous'
    else:
        return 'Others'
```

```python
df['tip_quality'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1)
df['tip_quality'].unique()
```
